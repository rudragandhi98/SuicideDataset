## Load the Data 

```python
# Loading the needed packages 
import pandas as pd 
import numpy as np

df = pd.read_excel('master.xlsx')
df.head()
```

## Change the Header row 

```python
df.rename(columns=df.iloc[0], inplace = True)
df.head()
```

## Remove the First Row 

```python
df = df.iloc[1:]
df.head()
```

## Drop Columns

```python
to_drop = ['country-year']
df.drop(to_drop, inplace=True, axis=1)
df.head()
```

## Fill in Missing Data for Columns

```python
df['HDI for year'].fillna((df['HDI for year'].mean()), inplace=True)
df.head()
```

## Create a New CSV file with the Clean Data

```python
# Load the clean data into a new csv file 
df.to_csv('Clean-Suicide-Data.csv')
```
